Input function (), this function is by default a string:

User weight (in pounds), then convert it to kilograms & print on terminal.

```python
weight_lbs = input("Weight in pounds: ")
weight_kg = int(weight_lbs) * 0.45
print(weight_kg) #This is the converted weight.
```

Coverting the US dollars to rands using the input() function:

```python
currency_dollar = input("Currency in dollar: ")
currency_rands = int(currency_dollar) * 17
print(currency_rands)
```

### Formatted strings:
The anatomy of a formatted string is f" {}", this allows user to change the data type to string.

```python
age = 35
age_formatted = f" I'm Mohau and {age} years old."
print(age_formatted)
```

```python
#More examples

temp = 18
txt_temp = f"It was so cold today, the weather was {temp} degrees maximum."
txt_temp
```
