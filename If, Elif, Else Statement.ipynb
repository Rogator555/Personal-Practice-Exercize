### If Statement:

```python
a = 11
b = 4

if a > b:
    print("a is bigger than b")

elif a == b:
    print("a is equal to b")

else:
    print("b is bigger than a")
```

Checking for negative or positive numbers:

```python
number = 23
if number < 0:
    print("number is negative")
else:
    print("number is positive")
```

Using if statement with an input() function by default is a String prompting a user to input infomation.

```python
# If statement with an input function() - Checking Age:

age = int(input())

if age <= 17:
    print("You're a minor")
elif age < 60:
    print("You're an adult")
else:
    print("You're a senior citizen")
```

```python
# Checking the weather:

weather = int(input("Temp readings: "))

if weather <= 18:
    print("Cold temperatures.")
elif weather < 26:
    print("Moderate temperatures.")
else:
    print("Hot temperatures.")
```
